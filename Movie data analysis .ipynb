```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("ggplot")
from matplotlib.pyplot import figure
%matplotlib.inline
import seaborn as sns
```

```python
# Read the data
data = pd.read_csv("movies.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
## filling na values
data['budget'].fillna(data["budget"].mean(),inplace = True)
```

```python
data.dropna(inplace = True)
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data["budget"] = data["budget"].astype('int64')
```

```python
data["gross"] = data["gross"].astype('int64')
```

```python
data.dtypes
```

```python
data.head()
```

```python
data.sort_values(by=["gross"],inplace = False,ascending = False)
```

```python
data.duplicated().sum()
```

```python
data
```

```python
### correlation with gross
### budget has high correlation
### company has high correlation
```

```python
#scatter plot with budget vs gross
plt.scatter(x = data["budget"],y = data["gross"])
plt.title("Budget Vs Gross")
plt.xlabel("Budget")
plt.ylabel("Gross")
plt.show()
```

```python
#plot the budget vs gross using seaborn
import seaborn as sns
sns.regplot(x = "budget",y="gross",data=data,scatter_kws ={"color":"red"},line_kws ={"color":"blue"})
```

```python
## correlation 
data.corr() #pearson, kendall, sparman
```

```python
# conclusion: high correlation between budget and gross
```

```python
cor_matrix = data.corr()
sns.heatmap(cor_matrix,annot = True)
plt.title("correaltion for numeric features")
plt.xlabel("movie features")
plt.ylabel("movie features")
plt.show()
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
data["company"] = le.fit_transform(data['company'])
```

```python
data_numerized = data

for col_name in data_numerized:
    if (data_numerized[col_name].dtype == "object"):
        data_numerized[col_name] = data_numerized[col_name].astype('category')
        data_numerized[col_name] = data_numerized[col_name].cat.codes
        
```

```python
data_numerized
```

```python
from matplotlib.pyplot import figure
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,8)
```
